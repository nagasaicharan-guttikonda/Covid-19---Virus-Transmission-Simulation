```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
import numpy as np
raw_data = []
with open('SEIR.csv') as File:
    reader = csv.DictReader(File)
    for row in reader:
        tp = type(row)
        raw_data.append(row)

res = []
for n in range(len(raw_data)):
    lst=list(raw_data[n].values())
    res.append(lst)
# print(res[0])

df = pd.DataFrame(res, columns = ['Days', 'Susceptible', 'Exposed', 'Infected', 'Recovered', 'Dead', 'R', 'K'])

# x axis - days
x =[]
strx = df["Days"].tolist()
for i in strx:
    x.append(int(i))

# y axis - people
# y_Sus stand for people of Susceptible
y_Sus =[]
strSus = df["Susceptible"].tolist()
for i in strSus:
    y_Sus.append(int(i))

# y_Ex stand for people of Exposed
y_Ex = []
strEx = df["Exposed"].tolist()
for i in strEx:
    y_Ex.append(int(i))

# y_Inf stand for people of Exposed
y_Inf = []
strInf = df["Infected"].tolist()
for i in strInf:
    y_Inf.append(int(i))

# y_Rec stand for people of Exposed
y_Rec = []
strRec = df["Recovered"].tolist()
for i in strRec:
    y_Rec.append(int(i))

# y_Dead stand for people of Exposed
y_Dead = []
strDead = df["Dead"].tolist()
for i in strDead:
    y_Dead.append(int(i))

fig,ax1 = plt.subplots()

plt.xlabel('days')
plt.ylabel('people')


width = 1



ax1.bar(x, y_Ex, width, label='Exposed')
ax1.bar(x, y_Inf, width, bottom = y_Ex, label='Infected')
ax1.bar(x, y_Sus, width, bottom = np.array(y_Inf) + np.array(y_Ex),
       label='Susceptible')
ax1.bar(x, y_Rec, width, bottom = np.array(y_Sus)+np.array(y_Inf) + np.array(y_Ex),
       label='Recovery')
ax1.bar(x, y_Dead, width, bottom = np.array(y_Rec)+np.array(y_Sus)+ np.array(y_Inf) + np.array(y_Ex),
       label='Dead')
plt.legend()
plt.savefig('Covid19_Simulation.jpg', dpi=300, bbox_inches='tight')
plt.show()
```
